```python
# test_qdrant_search.py (팀원과 함께 실행)

from agents.tools import search_sector_news_qdrant

def test_search():
    """Qdrant 검색 기능 테스트"""
    
    test_sectors = ["반도체", "바이오", "방위산업"]
    
    for sector in test_sectors:
        print(f"\n{'='*50}")
        print(f"테스트 섹터: {sector}")
        print('='*50)
        
        result = search_sector_news_qdrant.invoke({"sector_name": sector})
        
        print(f"총 결과: {result['total_results']}개")
        print(f"감성 통계: {result['sentiment_stats']}")
        print("\n상위 3개 뉴스:")
        for i, news in enumerate(result['news'][:3], 1):
            print(f"  {i}. [{news['sentiment']}] {news['title']}")
            print(f"     출처: {news['source']} | 점수: {news['score']}")

if __name__ == "__main__":
    test_search()
```
